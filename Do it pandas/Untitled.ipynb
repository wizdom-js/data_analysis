```python
import pandas as pd

drink_not = pd.read_excel('/Users/seuli/Downloads/stat_423801.xls', skiprows=2,encoding = "cp949")
drinks = drink_not.drop([15,16,17,18], axis = 0 )
print(drinks)
```

```python
male = drinks[2:8].fillna('남자')
female = drinks[9:].fillna('여자')
# drink.ffill(axis=0)
```

```python
drink = pd.concat([male, female])
print(drink)
```

```python
drink.columns = ['성별','연령대','2013','2014','2015','2016','2017','2018']
print(drink)
```

```python
drink_1 = pd.melt(drink, id_vars=['성별', '연령대'], var_name='연도', value_name='폭음률(%)')
drink_1
```
